# Data loading

```python
file_ped = '/home/martin/adapative_MC_test/parkinsons/parkinsons.ped'
file_map = '/home/martin/adapative_MC_test/parkinsons/parkinsons.map'
file_nosex = '/home/martin/adapative_MC_test/parkinsons/parkinsons.nosex'
output_file = '/home/martin/adapative_MC_test/parkinsons/parkinsons.pickle'
exclude_list = ['ND-5074', 'ND-5416', 'ND-5841']
df_nosex = pd.read_csv(file_nosex, delimiter='\t', header=None)
df_map = pd.read_csv(file_map, delimiter='\t', 
                     names=['chromosome', 'snp', 'start', 'end'])
n_snp = df_map.shape[0]
n_sample = df_nosex.shape[0]-3
genotype = np.zeros([n_sample, n_snp], dtype=int)
phenotype = np.zeros([n_sample], dtype=int)
temp_dic = {'0':0, 'A':1, 'T':2, 'C':4, 'G':8}
```

```python
i_sample = 0
with open(file_ped, 'r') as f:
    for line in f:
        line = line.strip().split(' ')
        sample_id = line[0]
        print(i_sample, sample_id)
        if sample_id not in exclude_list:
            phenotype[i_sample] = int(line[1])
            for i_snp in range(n_snp):
                genotype[i_sample, i_snp] = temp_dic[line[2*i_snp+2]] + temp_dic[line[2*i_snp+3]]
            i_sample = i_sample+1
```

```python
X = np.zeros([n_sample, n_snp], dtype=np.uint8)
for i in range(n_snp):
    unique_list, counts = np.unique(genotype[:,i], return_counts=True)
    ind_sort = np.argsort(counts)[::-1]
    unique_list_sort = unique_list[ind_sort]
    for i_unique,unique in enumerate(unique_list_sort):
        X[genotype[:,i]==unique, i] = i_unique
```

```python
prop_miss = np.mean(genotype==0, axis=0)
```

```python
# save the data 
with open(output_file, 'wb') as f:
    pickle.dump(X, f)
    pickle.dump(phenotype, f)
    pickle.dump(prop_miss, f)
```
